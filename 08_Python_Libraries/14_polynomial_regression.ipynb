```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = 6 * np.random.rand(100, 1) - 3
Y = 0.5 * X**2 + 1.5*X + np.random.rand(100,1)
```

```python
print(X)
```

```python
print(Y)
```

```python
plt.scatter(X, Y, color='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg
```

```python
lin_reg.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
score= r2_score(Y_test, lin_reg.predict(X_test))
print('R2 score for linear regression:', score)
```

```python
## Visualizing this model
plt.plot(X_train, lin_reg.predict(X_train), color = 'r')
plt.scatter(X_train, Y_train, color= 'b')
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
## Visualizing this model
plt.plot(X_test, lin_reg.predict(X_test), color = 'r')
plt.scatter(X_test, Y_test, color= 'b')
plt.xlabel("x axis")
plt.ylabel("y axis")
```

### Polynomial Linear Regression
Powerful extension of simple linear regression that allows us to capture more complex relationships between variables by introducing polynomial terms.

image.png

```python
## Applying Polynomial Transformation
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform (X_test)
```

```python
X_train_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_train_poly, Y_train)
y_pred = lin_reg2.predict(X_test_poly)
score = r2_score(Y_test, y_pred)
print('R2 score for polynomial Regression:', score)
```

```python
print('Coeficients:', lin_reg2.coef_)
print('Intercept:', lin_reg2.intercept_)
```

```python
plt.scatter(X_train, lin_reg2.predict(X_train_poly), color = 'r')
plt.scatter(X_train, Y_train, color = 'g')
```

```python
# Prediction of new dataset
X_new = np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
X_new_poly
```

```python
y_new = lin_reg2.predict(X_new_poly)
plt.plot(X_new, y_new, color = 'r', label = 'Predictions')
plt.plot(X_train, Y_train, 'g.', label= 'Training data')
plt.plot(X_test, Y_test, 'b.', label = 'Test data')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
plt.legend()
plt.show()
```

**Pipelining**

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
  X_new = np.linspace(-3, 3, 200).reshape(200, 1)
  poly_features = PolynomialFeatures(degree = degree)
  lin_reg = LinearRegression()
  poly_regression = Pipeline([
      ('poly_features', poly_features),
      ('lin_reg', lin_reg)
  ])
  poly_regression.fit(X_train, Y_train)
  y_pred_new = poly_regression.predict(X_new)
  # Plotting prediction line
  plt.plot(X_new, y_pred_new, label = 'predictions', color='r')
  plt.plot(X_train, Y_train, 'g.', label = 'Training data')
  plt.plot(X_test, Y_test, 'b.', label = 'Test data')
  plt.legend(loc = 'upper left')
  plt.xlabel('x dataset')
  plt.ylabel('y dataset')
  plt.show()
```
