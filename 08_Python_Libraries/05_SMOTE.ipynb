# SMOTE (Synthetic Minority Oversampling Technique)
Technique used in machine learning to address imbalanced datasets where the minority class has the significantly fewer than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.


```python
from sklearn.datasets import make_classification
```

```python
# x is independant feature and y is dependant feature
x,y = make_classification(n_samples = 1000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights = [0.90], random_state=1)
```

```python
import pandas as pd

df1 = pd.DataFrame(x, columns= ['f1', 'f2'])
df2 = pd.DataFrame(y,columns = ['target'])
final_df = pd.concat([df1,df2], axis=1)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x,y = oversample.fit_resample(final_df[['f1', 'f2']],final_df['target'])
```
