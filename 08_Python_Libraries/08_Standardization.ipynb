```python
import numpy as np
```

```python
total_bill = list(df['total_bill'])
```

```python
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean, std
```

```python
data = []
for i in total_bill:
  z_score = (i - mean)/std
  data.append(z_score)
```

```python
data
```

```python
max(data)
```

```python
min(data)
```

```python
sns.histplot(total_bill, kde=True)
```

```python
sns.histplot(data,kde=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
df.head()
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
scaler.transform(df[['total_bill', 'tip']])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(scaler.transform(df[['total_bill', 'tip']]), columns=['total_bill_standard', 'tip'])
```
