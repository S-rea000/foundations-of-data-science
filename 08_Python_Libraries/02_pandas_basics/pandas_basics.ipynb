# Pandas
Pandas library in python is used for cleaning, exploring, manipulating & working with the datasets.

#### Two primary data structures used for data manipulation & analysis:
* Series 
* Dataframe

Series is a 1D labelled array that best combines list & dictionary, maintains single collection of orderedvalue, value is assigned with identifier, doesnot have to be unique.

```python
import pandas as pd # import & installation is a prime task before using any libraries
```

```python
icecream = ['Chocolate', 'Vanilla', 'Strawberry', 'Butterscotch']
print(pd.Series(icecream))
```

```python
# Create a series object from a dictionary
fish = {       
    'Salmon': 'Orange',  # dictionary key is used as index
    'Tuna': 'Red',
    'Eel': 'Brown'
}
pd.Series(fish)
```

```python
# Intro to Series method
prices = pd.Series([2.99, 4.45, 1.36])
prices.sum()
```

```python
prices * 3
```

```python
prices.product()
```

```python
prices.mean()
```

```python
prices.std()
```

```python
# Intro to attribute: Any information related to data
# size: total data in series
# is_unique: duplicated values sequence
adjectives = pd.Series(['Smart', 'Charming', 'Brilliant', 'Humble', 'Smart'])
print(adjectives)
```

```python
adjectives.size
```

```python
adjectives.is_unique
```

```python
# Parameters & Arguments
fruits = ['Apple', 'Orange', 'Plum', 'Grapes', 'Watermelon']
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
pd.Series(fruits, weekdays)
```
