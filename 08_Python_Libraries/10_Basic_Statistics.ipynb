#**Introduction to Statistics**
Statistics is a science of collecting, organizing, & analyzing data.
Data is a fact or piece of information.

**Types of Statistics**

**1. Descriptive Statistics:**
* Obtaining information from existing data's
* Mostly used by data analyst
* Organizing and summarizing data.
* Measure of central tendency &dispersion.

 **2. Inferential Statistics**
 * Using collected data to form conslusion.
 * Mostly used by data scientist
 * Help making decisions
 * Z test, t test, Chi square test

**Spearman's Rank Correlation**
Measures strength and direction of association between two variables based on their ranks instead of actual value.

```python
import seaborn as sns
df = sns.load_dataset('healthexp')
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
df.corr(numeric_only= True, method='pearson')
```

```python
df.select_dtypes(exclude= ['number'])
```

```python
df.cov(numeric_only=True)
```

```python
df.corr(numeric_only=True, method= 'spearman')
```

```python
sns.heatmap(df.corr(numeric_only=True, method='pearson'), annot= True, cmap='magma')
```
