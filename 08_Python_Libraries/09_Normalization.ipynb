# Min-max scalar(Normalization)
Mostly used in deep learning. Min-max formula is used.
image.png

```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
df[['distance', 'fare', 'tip']]
```

```python
min(df['distance']), max(df['distance']), min(df['fare']), max(df['fare']),min(df['tip']), max(df['tip'])
```

```python
min_max.fit(df[['distance', 'fare', 'tip']])
```

```python
min_max.transform(df[['distance', 'fare','tip']])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(min_max.fit_transform(df[['distance','fare','tip']]), columns = ['distance', 'fare', 'tip'])
```

```python
min(df['distance']), max(df['distance']), min(df['fare']), max(df['fare']),min(df['tip']), max(df['tip'])
```
