# Data Encoding:
Process of converting categorical or textual data into numerical format that machine learning algorithms can understand & process.
Types of data encoding:
* One hot encoding
* Label Encoding
* Target guided Ordinal Encoding


```python
# 1. Nominal or OHE (No specific order or rank)
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color': ['red', 'blue', 'green','green','blue', 'red']
})
```

```python
df
```

```python
type(df)
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(df[['color']])
```

```python
encoded = ohe.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python
!pip install -U scikit-learn
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out(['color'])
)
```

```python
encoded_df
```

```python
df
```

```python
pd.concat([df, encoded_df], axis = 1)
```

Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoder.fit_transform(df[['size']])
```
