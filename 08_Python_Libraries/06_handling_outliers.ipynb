# Handling Outliers
* Outlier: Data that's completely different from the normal one. Impacts the machine learning algorithms which decreases the performance and accuracy of algorithms.
* Five number summary: Minimum Value - Lower quartile(Q1 25%) - Median(Q2 50%) - Upper quartile(Q3 75%) - Maximum Value

```python
import numpy as np
```

```python
# list of nubers with outliers
numbers = [45, 32, 24, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74, 1000, 1100]
```

```python
np.percentile(numbers, 0) # minimum value
```

```python
np.percentile(numbers, 25) #lower quartile
```

```python
np.percentile(numbers, 50) # median
```

```python
np.percentile(numbers, 75) #upper quartile
```

```python
np.percentile(numbers, 100) # maximum value
```

```python
# Finding Outliers
# [lower fence <--> Upper fence]
minimum, Q1, median, Q3, maximum = np.quantile(numbers,[0, 0.25,0.5, 0.75, 1])
```

```python
minimum
```

```python
Q3
```

```python
maximum
```

```python
iqr = Q3 - Q1
print(iqr)
```

```python
lower_fence= Q1 - 1.5 *iqr
higher_fence = Q3 + 1.5 * iqr
```

```python
lower_fence, higher_fence
```

```python
outliers = []
for num in numbers:
  if num < lower_fence or num > higher_fence:
    outliers.append(num)
```

```python
print (outliers)
```
