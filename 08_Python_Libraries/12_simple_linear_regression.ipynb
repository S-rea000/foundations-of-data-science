# Simple Linear Regression
Statistical method to model relationship between one dependent variable(predictor) & one independent variable(response)

Goal is to find the best fit line in order to minimize error.
* Cost Function: Avrage squared distance between actual value and predicted value for a dataset. (Used to find best fit line)
Minimum value = best fit line
image.png

Types of Cost function:
1. Mean Squared Error(MSE): Big mistakes should be punished more.
2. Mean Absolute Error(MAE): every error is treated equally.
3. Root Mean Squared Error(RMSE): when values should be used as same unit.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title(' Tip Vs Total Bill')
plt.show()
```

```python
X = df[['total_bill']] #Independent feature
Y = df[['tip']] # Dependent Feature
```

```python
X, Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42)
```

```python
X.shape
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
plt.scatter(X_train, Y_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
print('The slope or coefficient of weight is:', model.coef_)
```

```python
print('The intercept of the line is:', model.intercept_)
```

```python
plt.scatter(X_train, Y_train)
plt.plot(X_train, model.predict(X_train), 'r')
```

```python
plt.scatter(X_test, Y_test)
```
