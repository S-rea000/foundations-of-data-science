## Basic Conditional Statements


**Even or Odd Checker**

```python
num = int(input('Enter a number:'))
if num % 2 == 0:
  print('Even')
else:
  print('Odd')
```

**Positive, Negative, or Zero**

```python
num = int(input('Enter a number:'))
if num < 0:
  print('This number is negative')
elif num > 0:
  print ('This number is positive')
else:
  print ('This number is zero')
```

**Smallest of Three Numbers**

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if a > b < c:
  print (f'{b} is the smallest number')
elif a > c < b:
  print (f'{c} is the smallest number')
else:
  print (f'{a} is the smallest number')
```

**Grade Categorizer**

```python
score= int (input('Enter your score:'))
if score < 0:
  print ('Score should not be negative')
elif score > 100:
  print ('Score should not be greater then 100')
else:
  if score >= 90:
    print ('Grade A')
  elif score >= 80 and score < 90:
    print ('Grade B')
  elif score>= 70 and score < 80:
    print ('Grade C')
  else:
    print('Failed')
```

**Simple Password Check**

```python
password = 'secret'
pwd = input('Enter a password:')
if pwd == password:
  print ('Access granted')
else:
  print ('Access denied')
```

## Conditional Logic with Loops


**Multiples of 3 and 5**

```python
n = int (input("Enter a single integer: "))
for i in range(1, n+1):
  print (i)
  if i % 3 == 0 and i % 5 == 0:
    print('Multiple of both')
  elif i % 3 == 0:
      print ('Multiple of 3')
  elif i % 5 == 0:
      print ('Multiple of 5')
  else:
    print(f"{i} is not a multiple of 3 or 5")
```

**Categorize scores froma list**

```python
scores = [85, 42, 78, 99, 65]
for i in scores:
  print (i)
  if i >= 60:
    print("Pass")
  else:
    print("Fail")
```

## Looping Over Numeric Ranges


**Print Numbers from 1 to 10**

```python
for i in range (1, 11):
  print(i)
```

**Print Even Numbers Between 1 and 50**

```python
for i in range (1, 51):
  if i % 2 == 0:
    print(i)
  else:
    print("")
```

## Looping Over Strings


**Print All Characters in a String**

```python
st1 = input("Enter a word: ")
for i in st1:
  print (i)
```
