**Operator Precedence**

```python
a = 4
b = 2
c = 5
```

```python
a + b * c
```

```python
(a + b) * c
```

In this section of Task 3, we can see that;
*   When there's an absence of operator precedence, it simply follows the BODMAS rule, multiples b with c (5 * 2 = 10), and then adds the result with a (10 + 4 = 14).
*   When there's a presence of operator precedence, it priotorizes the the precendence part i.e., (a + b) is priotorized (4 + 2 = 6) and then * ( 6 * 5 = 30)
