Iterate Through a Tuple and Print Types

```python
t = (False, 1, 2.3, True, 5, 'Shreya', 8-1j, 'Eight',9, 1-5j)
```

```python
for i in t:
  print(i)
  print(type(i))
```

Print All Items from a List with a Custom Message

```python
l = ['Thimi', 'Lalitpur', 'Kathmandu','Pokhara']
```

```python
for i in l:
  print (f'{i} is a great place')
```

Print Each Character of a String with Its Index

```python
input1 = input('Enter a word: ')
for index, i in enumerate(input1):
  print(f'{index}: {i}')
```

Sum of Elements in a List

```python
l1 = [1, 5, 7, 2, 8, 4]
total_sum = 0

for i in l1:
  total_sum += i
  print(f'The given numbers are {i}')
print (f'The total sum is {total_sum}')
```

Count Booleans in a Tuple

```python
t1 = ('Apple', 0, 3-7j, 1, 2, True, False, 7, 9, True )
true_count = 0
for i in t1:
  if i is True:
    true_count += 1
print(f'The number of true values in the tuple is {true_count}')
```

Check and Print Data Types from a List

```python
k = [9, 'hey', 0, True, 7-1j, 78, False, 'end']
for i in k:
  print (f'{i} is a {type(i)}')
```

Check for Vowels in a String

```python
user_input = input('Enter a word: ')
vowel = ('a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U')
for i in user_input:
  if i in vowel:
    print(f'{i} is a vowel')
```

Print Square of Numbers from a Tuple

```python
t2 = (1, 2, 3, 4, 5)
square = ()
for i in t2:
  square = i**2
  print(f'The square of {i} is {square}')
```

Print Elements of a List in Uppercase

```python
s = ['apple', 'tree', 'monkey', 'ate', 'stomach']
upper_s = ()
for i in s:
  upper_s = i.upper()
  print(upper_s)
```

Count Numbers Greater Than 10 in a List

```python
integers = [3, 5, 9, 10, 15, 20, 25]
integer_count = 0
for i in integers:
  if i < 10:
    integer_count += 1
print (f'Count of integers greater than 10 is {integer_count} ')
```
