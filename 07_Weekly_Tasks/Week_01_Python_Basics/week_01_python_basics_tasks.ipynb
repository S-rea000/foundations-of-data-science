**12. Arithmetic Operator Precedence**

```python
a = 5
b = 3
c = 2
```

```python
a + b * c
```

```python
(a + b) * c
```

In this section of Task 1, we can see that;
*   When there's an absence of operator precedence, it simply follows the BODMAS rule, multiples b with c (3 * 2 = 6), and then adds the result with a (5 + 6 = 11).
*   When there's a presence of operator precedence, it priotorizes the the precendence part i.e., (a + b) is priotorized (5 + 3 = 8) and then * ( 8 * 2 = 16)
