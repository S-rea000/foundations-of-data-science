# Encapsulation:
Encapsulation is a core pillar of Object-Oriented Programming (OOP) that involves bundling data (attributes) and the methods that operate on that data into a single unit, typically a class. It serves as a protective shield that prevents data from being accessed or modified directly by code outside that unit, ensuring data integrity and security.

 ### Access Modifier
* Public:	member	-> Accessible from anywhere, both inside and outside the class.
* Protected:	_member ->	A convention indicating the member is for internal use by the class and its subclasses.
* Private:	__member -> Restricts access to the class itself using name mangling to prevent accidental overrides.

```python
class bank_account:
  def __init__(self, balance):
    self.__balance = balance

  def deposit (self, amount):
    if amount <= 0:
      return 'Invalid Deposit'
    self.__balance = self.__balance + amount
    return f'Rs.{amount} deposited & final balance after the deposit is Rs.{self.__balance}'

  def withdraw (self, amount):
    if self.__balance >= amount:
      self.__balance= self.__balance - amount
      return f'Rs.{amount} has been withdrawn from your account. The remaining balance is Rs.{self.__balance}'
    else:
      return f'Invalid withdrawl! Not ENOUGH BALANCE!!'

  def get_balance(self):
    return f'The total balance in your account is Rs.{self.__balance}'
```

```python
Shreya = bank_account(5000)
```

```python
Shreya.deposit(500)
```

```python
Shreya.withdraw(4500)
```

```python
Shreya.withdraw(1500)
```

```python
Shreya.get_balance()
```

### Key Benefits
* Security: Prevents unauthorized or accidental modification of critical variables.

* Flexibility: Internal implementation can be changed (e.g., changing __balance from a float to a decimal) without breaking external code that relies on get_balance().
