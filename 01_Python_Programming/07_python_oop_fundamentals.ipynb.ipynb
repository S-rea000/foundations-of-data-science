## Instance Variables vs Class Variables

In Python, variables inside a class can either belong to:

- The **instance (object)** → Instance Variable  
- The **class itself** → Class Variable  

### Instance Variables
- Defined inside the constructor (`__init__`)
- Unique to each object
- Accessed using `self.variable_name`

### Class Variables
- Defined directly inside the class (outside methods)
- Shared by all instances of the class
- Accessed using `ClassName.variable_name`


```python
class Student:
    # Class variable (shared by all objects)
    school_name = "Green Valley School"

    def __init__(self, name, grade):
        # Instance variables (unique per object)
        self.name = name
        self.grade = grade


# Creating objects
student1 = Student("Alice", 10)
student2 = Student("Bob", 12)

# Accessing instance variables
print(student1.name)      # Alice
print(student2.grade)     # 12

# Accessing class variable
print(Student.school_name)      # Green Valley School
print(student1.school_name)     # Green Valley School
```

```python
# Modifying class variable
Student.school_name = "Blue Ridge School"

print(student1.school_name)
print(student2.school_name)

# Creating an instance variable with same name (overrides only for that object)
student1.school_name = "Private School"

print(student1.school_name)  # Private School
print(student2.school_name)  # Blue Ridge School
```

### Key Difference

- Changing a **class variable** affects all instances.
- Changing an **instance variable** affects only that specific object.
- If an instance variable has the same name as a class variable, it overrides the class variable for that object only.
