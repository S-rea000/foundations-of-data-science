```python
def number (a):
  l1 = []
  for i in a:
    if type(i) == list:
      for j in i:
        if type (j) == int or type(j) == float:
          l1.append(j)
    else:
      if type(i) == int or type(i) == float:
        l1.append(i)
  return l1
```

**Lambda Function**:

* Small, one-line function in python.
* function defined is used as variable.

```python
def pow(n,p):
  return n ** p
```

```python
pow (6, 2)
```

Format:

func = lambda inputs: returning expression

```python
power = lambda n,p: n**p
```

```python
print(f"The power is {power (5, 2)}")
```

```python
add = lambda x,y: x + y
```

```python
add (7, 3)
```

```python
subtract = lambda a,b = 4: a - b # To keep a default value
```

Map Reduce Function:

* Map function: Map datas in a function.

```python
l = [1,2,3,4,5]
```

```python
def square(a):
  l1 = []
  for i in a:
    l1.append(i**2)
  return l1
```

```python
square(l)
```

```python
def square (x):
  return x**2
```

```python
map(square,l)
```

```python
list(map(square,l))
```

```python
strings = ['Shreya','Hello','Great']
```

```python
list(map(len,strings))
```

```python
frozenset(map(square,l))
```

```python
tuple(map(lambda x : x**3,l))
```

```python
l1 = [1, 2, 3, 4, 5]
l2 = [5, 6, 7, 8, 9]
list(map(lambda x,y: x+y, l1,l2))
```

```python
s = 'lalitpur'
list(map(lambda a : a.upper(),s))
```
