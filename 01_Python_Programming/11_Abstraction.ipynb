# Abstraction:
Abstraction is an Object-Oriented Programming (OOP) concept that focuses on hiding the complex, underlying implementation details of a system and only exposing the essential features to the user. It provides a blueprint for other classes to follow.

In Python, abstraction is strictly implemented using the abc (Abstract Base Classes) module.

**Key Rules of Abstraction:**

* Cannot be instantiated: You cannot create an object directly from an Abstract Base Class.

* Strict Enforcement: Any child class inheriting from an ABC must provide concrete implementations for all methods marked with @abstractmethod.

```python
def circle_area(radius):
  return 3.14*radius*radius
print("Area of Circle is: ", circle_area(6))
```

```python
def area(radius):
  return round(3.14*radius*radius, 2)
print(area(2))
```

```python
import abc # Module for abstraction
class taylors:
  @abc.abstractmethod
  def student_details(self):
    pass
  @abc.abstractmethod
  def student_assignment(self):
    pass
  @abc.abstractmethod
  def student_marks(self):
    pass
```

```python
class student_details(taylors):
  def student_details(self):
    return 'This is the method for taking student details'
  def student_assignment(self):
    return 'This is a method for assignments specially for students'
class data_analysis(taylors):
  def student_details(self):
    return 'This will return the details for data analysis course'
  def student_assignment(self):
    return 'This will give assignment details'

```

### Procedural vs. OOP Abstraction
A simple function hides complex math (procedural abstraction), but OOP abstraction forces a consistent structure across different objects.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    """Abstract Base Class acting as a blueprint for all shapes."""
    
    @abstractmethod
    def calculate_area(self) -> float:
        """Child classes must implement this method."""
        pass

class Circle(Shape):
    def __init__(self, radius: float):
        self.radius = radius
        
    def calculate_area(self) -> float:
        # Concrete implementation of the abstract method
        return round(math.pi * (self.radius ** 2), 2)

class Rectangle(Shape):
    def __init__(self, length: float, width: float):
        self.length = length
        self.width = width
        
    def calculate_area(self) -> float:
        # Concrete implementation of the abstract method
        return float(self.length * self.width)

# Usage
circle = Circle(6)
rect = Rectangle(4, 5)

print(f"Circle Area: {circle.calculate_area()}")    
print(f"Rectangle Area: {rect.calculate_area()}")   

# Shape()  <-- This would raise a TypeError because Shape is abstract.
```

```python
from abc import ABC, abstractmethod

class UniversityCourse(ABC):
    """Abstract Base Class for all university courses."""
    
    @abstractmethod
    def get_syllabus(self) -> str:
        pass
        
    @abstractmethod
    def get_assignment_criteria(self) -> str:
        pass
        
    @abstractmethod
    def calculate_final_grade(self) -> str:
        pass

# ---------------------------------------------------------
# Concrete Implementations
# ---------------------------------------------------------

class DataAnalysisCourse(UniversityCourse):
    """A concrete class representing a specific course."""
    
    def get_syllabus(self) -> str:
        return "Syllabus: Python, Pandas, SQL, and Data Visualization."
        
    def get_assignment_criteria(self) -> str:
        return "Criteria: 3 Jupyter Notebook projects and 1 final presentation."
        
    def calculate_final_grade(self) -> str:
        return "Grading: 60% Projects, 40% Final Exam."

class SoftwareEngineeringCourse(UniversityCourse):
    """Another concrete class strictly following the UniversityCourse blueprint."""
    
    def get_syllabus(self) -> str:
        return "Syllabus: OOP, System Design, Git, and CI/CD."
        
    def get_assignment_criteria(self) -> str:
        return "Criteria: Weekly coding labs and 1 group software project."
        
    def calculate_final_grade(self) -> str:
        return "Grading: 50% Group Project, 50% Individual Coding Assessments."

# ---------------------------------------------------------
# Usage
# ---------------------------------------------------------

da_course = DataAnalysisCourse()
se_course = SoftwareEngineeringCourse()

print(da_course.get_syllabus())
print(se_course.get_assignment_criteria())
```
