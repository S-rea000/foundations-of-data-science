# Python Comprehensions

This notebook demonstrates the use of list, dictionary, and set comprehensions in Python for writing concise and readable code.


## Comprehension
* Shortcut way to create lists, sets, or dictionaries in python.
* Helps make lists quickly

### Types:
* List Comprehension: [Expression for item in iterable if condition.]
* Set Comprehension: (Exprehension for item in iterable if condition.)
* Dictionary Comprehension: {key_expression: value_expression for item in iterable if condition}
* Generator Comprehension: (expression for item in iterable if condition)

```python
# Normal format:
l = [1, 2, 3, 4, 5]
l1 = []
for i in l:
  l1.append(i**2)
l1
```

```python
# When using comprehension
l1 = [i ** 2 for i in l]
l1
```

```python
[i for  i in l if i % 2 == 0] # Even
```

```python
[i for i in l if i % 2 == 1] # Odd
```

```python
[i for i in l if i % 2 != 0] # Odd
```

```python
[i for  i in l]
```

```python
l1 = ['this', 'place', 'provides', 'internships']

# List Comprehension
[i.upper() for i in l1]
```

```python
tuple (i for i in l1)
```

```python
d = {
    'key1': 1,
    'key2': 2,
    'key3': 3,
    'key4': 4
}
```

```python
d.items()
```

```python
result = {}
for k,v in d.items():
  result[k] = v**2
print(result)
```

```python
{k: v**2 for k,v in d.items()}
```

## Why Use Comprehensions?

Comprehensions provide a concise and readable way to create collections compared to traditional loops.
