# File Handling
File handling is a crucial part of software engineering and data science. It allows applications to persist data, read configurations, and parse datasets.

In Python, the most robust and professional way to handle files is by using a Context Manager (the with statement). Context managers ensure that resources are efficiently allocated and automatically closed after use, preventing data corruption and memory leaks.

**Common File Modes:**

* 'r': Read (default). Opens a file for reading.

* 'w': Write. Opens a file for writing (creates a new file or truncates an existing one).

* 'a': Append. Opens a file for writing, appending to the end of the file if it exists.

* 'b': Binary mode (e.g., 'wb', 'rb').

```python
# 1. WRITING to a file
f = open('test.txt', 'w') #creating a new file
```

```python
f.write('This is my first file. ')
```

```python
f.close()
```

```python
# 2. APPENDING to a file
f = open ('test.txt', 'a')
```

```python
f.write('This is my datascience learning journey')
```

```python
# 3. READING from a file
f = open('test.txt', 'r')
```

```python
f.read()
```

```python
f.readline()
```

```python
f.seek(5)
f.readline()
```

```python
for i in f:
  print (i)
```

```python
f.close()
```

### Working with JSON Data
JSON (JavaScript Object Notation) is the standard format for data exchange on the web and configuration files.

```python
data = {
    'name': 'Shreya',
    'email': 'srea@gmail.com',
    'phone.no': 1111111111
}
```

```python
import json

with open('data.json', 'w') as f:
    json.dump(data, f)   # saves data as JSON
```

```python
f.close()
```

```python
with open ('data.json', 'r') as f:
  data1 = json.load(f)
```

```python
data1
```

### Working with CSV Data
CSV (Comma Separated Values) is widely used for tabular data, especially in machine learning and data analysis

```python
import csv #comma separated value
data = [
    ['name', 'email_id','phone_number'],
    ['shreya','srea@gmail.com', 9999999],
    ['Riya','rea@gmail.com', 98989898],
     ['priya','pry@gmail.com', 9777999],
]
```

```python
with open ('data.csv', 'w') as f:
  writer = csv.writer(f)
for i in data:
  writer.writerow(i)
```

```python
with open('data.csv', 'r') as f:
    read_data = csv.reader(f)
    for i in read_data:
        print(i)
```
