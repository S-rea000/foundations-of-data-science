# Polymorphism in Python

This notebook explains the concept of polymorphism in Python with clear examples and practical demonstrations.


## What is Polymorphism?

Polymorphism means "many forms". In Python, it allows different objects to respond to the same method name in different ways.

* Example: fly() method: Flap wings for birds; Start engine for airplanes; Operate engine for car

## Example 1: Built-in Polymorphism

This example shows how the same function behaves differently for different data types.


```python
print(len("Hello"))
print(len([1, 2, 3, 4]))
```

```python
def add(a,b):
  return a + b
```

```python
add (3,5) # Sum
```

```python
add ('Data', 'Science') # Conncatenation
```

## Example 2: Polymorphism Using Functions

This example demonstrates how a single function can operate on different object types.


```python
def func(obj):
    obj.speak()

class Dog:
    def speak(self):
        print("Bark")

class Cat:
    def speak(self):
        print("Meow")

func(Dog())
func(Cat())
```

```python
class dataScience:
  def syllabus (self):
     print('Python', 'Pandas', 'Numpy', 'Seaborn', 'matplotlib', 'R', 'SQL')
```

```python
class mern:
  def syllabus (self):
     print('java', 'javascript', 'mongodb', 'nodejs', 'reactjs')
```

```python
def classes (class_obj):
  for i in class_obj:
    i.syllabus()
```

```python
dataScience = dataScience()
mern = mern()
obj = [dataScience, mern]
```

```python
classes(obj)
```

## Example 3: Polymorphism with Method Overriding

This example shows how child classes provide different implementations of the same method.


```python
class Bird:
    def fly(self):
        print("Some birds fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies")

class Ostrich(Bird):
    def fly(self):
        print("Ostrich cannot fly")
```

## Conclusion

Polymorphism improves flexibility and code reusability. It allows the same interface to behave differently depending on the object, which is a key principle of Object-Oriented Programming in Python.
