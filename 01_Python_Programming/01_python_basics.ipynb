Indexing and Slicing String

Indexing:

Finding the position/index. Indexing starts from the number 0,1,2,...

Positive indexing: Left to right || Negative indexing: Right to left

Format: Variable [index]

Positive indexing: n[0,1,2,3....]  || Negative indexing: n[-1,-2,-3,...]

```python
n = 'Indexing'
```

```python
n[7] #positive indexing
```

```python
n [-7]
```

Slicing:

Slicing the strings

Format: Variable[start:end]  || variable [start:end: step]

```python
m = "Slicing"
```

```python
m[0:4] # end is exclusive(excluded)
```

```python
m[4:8]
```

```python
place = "putalisadak"
```

```python
place[0:6]
```

```python
place [6:11]
```

```python
place [0:11:1] #default value of step is 1
```

```python
place[0:11:3] # can be written as place[0::3] or place [::3]
```

**DAY 3**

```python
place [0:10:-1] #Starting should always be less than the ending or else directional conflict arises giving a blank result
```

```python
place [-10:-4:] #Since -10 < -4 it shows no issues
```

```python
place [-10:-4:-2] # directional conflict || start must be greater than stop during negative slicing
```

```python
place [-4:-10:-2]
```

String Reverse

```python
place[::-1]
```

Reverse Step
