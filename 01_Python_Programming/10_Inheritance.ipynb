# Inheritance
Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class (the Child or Subclass) to acquire the properties and methods of another class (the Parent or Superclass). This promotes code reusability, establishes a natural hierarchy, and makes architectures much easier to maintain.

**Primary Types of Inheritance:**

* Single Inheritance: A child class inherits from a single parent class.

* Multi-Level Inheritance: A child class inherits from a parent, which in turn inherits from a grandparent class (a chain).

* Multiple Inheritance: A child class inherits from more than one parent class simultaneously.

### Single Inheritance
In Single Inheritance, one subclass derives from one superclass.

```python
# Simple Inheritance
class parent:
  def parent_method(self):
    return ' This is an example of single inheritance.'
```

```python
p = parent() # initializing class object
```

```python
p.parent_method()
```

```python
class child(parent):
  pass
```

```python
c = child()
```

```python
c.parent_method()
```

### Multi-Level Inheritance
In Multi-Level Inheritance, features of the base class and the derived class are inherited into a new derived class. Think of it as a Grandparent → Parent → Child relationship.

```python
class Grandparent:
    def trait_one(self) -> str:
        return "Trait 1: Wisdom (from Grandparent)"

class Parent(Grandparent):
    def trait_two(self) -> str:
        return "Trait 2: Hard Work (from Parent)"

class Child(Parent):
    def trait_three(self) -> str:
        return "Trait 3: Tech-Savvy (from Child)"

# Usage
# The Child object has access to its own methods AND the methods of all its ancestors.
child_obj = Child()
print(child_obj.trait_one())   # From Grandparent
print(child_obj.trait_two())   # From Parent
print(child_obj.trait_three()) # From Child
```

```python

class class1:
  def test_class1(self):
    return 'This is a method from class 1.'
class class2:
  def test_class2(self):
    return 'This is a method from class 2'
class class3(class2):
  pass
```

```python
obj_class1 = class1()
```

```python
obj_class1.test_class1()
```

```python
obj_class2 = class2()
```

```python
obj_class2.test_class2()
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class2()
```

### Multiple Inheritance
In Multiple Inheritance, a single derived class inherits from two or more base classes.

```python
class father:
  def patriarchial_property(self):
    return "Father's property inherited."
class mother:
  def matriarchial_property(self):
    return "Mother's property inherited."
class child (father, mother):
  pass
```

```python
obj_child = child()
```

```python
obj_child.patriarchial_property()
```

```python
obj_child.matriarchial_property()
```

### super() Function
In real-world applications, you often need to add new attributes to a child class while keeping the parent's initialization. We use super() to achieve this cleanly.

```python
class Employee:
    """Base class for all employees."""
    def __init__(self, name: str, salary: float):
        self.name = name
        self.salary = salary

class Manager(Employee):
    """Derived class that adds a specific attribute for Managers."""
    def __init__(self, name: str, salary: float, department: str):
        # super() calls the __init__ of the Employee (Parent) class
        super().__init__(name, salary)
        self.department = department

    def get_details(self) -> str:
        return f"{self.name} manages the {self.department} department and earns ${self.salary}."

# Usage
manager = Manager("Alice", 120000, "Engineering")
print(manager.get_details())
```
