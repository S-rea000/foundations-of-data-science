# Exception Handling:
Exception handling is a critical mechanism in software development that gracefully manages runtime errors. Instead of allowing a program to crash abruptly when an error occurs, exception handling catches the error and dictates how the program should respond.

The Four Pillars of Exception Handling:
* try: Contains the code that might raise an exception. 
* except: Defines the code to execute if an error occurs in the try block. 
* else: Executes only if no exceptions were raised in the try block. 
* finally: Executes unconditionally, regardless of whether an error occurred or not. Often used for cleanup. 


```python
a = 10
# a/0 --> Zero Division Error
```

```python
# How exception handling works
try:
  a/0
except Exception as e:
  print(e)
```

```python
try:
  f = open('test.txt', 'r') # --> Non-existent file
except Exception as e:
  print (e)
```

```python
try:
  f = open('test.txt', 'w')
  f.write('This document is about exception handling')
except exception as e:
  print(e)
else:
  print('The code has been executed successfully')
```

```python
try:
  f = open('test1.txt', 'w')
  f.write('write something')
  f.close()
except Exception as e:
  print('This is except block', e)
finally:
  print('This finally block would run in any situation')
```

```python
try:
  f = open('test1.txt', 'r')
  f.write('write something')
  f.close()
except Exception as e:
  print('This is except block', e)
finally:
  print('This finally block would run in any situation')
```

### Custom Exception:
Sometimes, built-in exceptions don't cover specific business logic rules. You can create custom exceptions by inheriting from Python's base Exception class and using the raise keyword.

```python
age = int(input('Enter your age: '))

class validateAge(Exception):
    def __init__(self, msg):
        self.msg = msg
    def validateAge(age):
      if age<0:
          raise validateAge('The input age is invalid')
      elif age > 150:
          raise validateAge('Too high to exist!')
      else:
          print('Validated age')
validateAge(age)
```
